```python
# 必要なライブラリのインポート
import pandas as pd
import numpy as np

# ダミーデータの作成
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace', 'Heidi', 'Ivan', 'Judy'],
    'Age': [25, 30, 35, 40, 45, 50, 55, 60, 65, 70],
    'Gender': ['F', 'M', 'M', 'M', 'F', 'M', 'F', 'F', 'M', 'F'],
    'Score': [88, 92, np.nan, 94, 78, np.nan, 85, 90, 95, 88]
}
df = pd.DataFrame(data)

# headメソッドの使用例
print("headメソッドの使用例:")
print(df.head())

# tailメソッドの使用例
print("\ntailメソッドの使用例:")
print(df.tail())

# uniqueメソッドの使用例
print("\nuniqueメソッドの使用例:")
print(df['Gender'].unique())

# nuniqueメソッドの使用例
print("\nnuniqueメソッドの使用例:")
print(df['Gender'].nunique())

# isnullメソッドの使用例
print("\nisnullメソッドの使用例:")
print(df['Score'].isnull())

# sort_valuesメソッドの使用例
print("\nsort_valuesメソッドの使用例:")
print(df.sort_values(by='Age'))

# groupbyメソッドの使用例
print("\ngroupbyメソッドの使用例:")
print(df.groupby('Gender').mean())

# queryメソッドの使用例
print("\nqueryメソッドの使用例:")
print(df.query('Age > 50'))

# value_countsメソッドの使用例
print("\nvalue_countsメソッドの使用例:")
print(df['Gender'].value_counts())

# applyメソッドの使用例
print("\napplyメソッドの使用例:")
print(df['Name'].apply(lambda x: x.upper()))
```
