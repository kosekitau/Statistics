```python
import numpy as np
import pandas as pd
import polars as pl
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
df = pd.read_csv('/content/drive/MyDrive/study/modern_marketing_research/13/apriori.txt', sep=',')
print(df.shape)
df.head()
```

```python
# 各商品たちの組み合わせの支持度を求める
# supportは支持度、confidenceは信頼度
result_support = apriori(df, min_support=0.4, use_colnames=True)
```

```python
result_support
```

```python
# アソシエーションルールの抽出
result_association = association_rules(result_support, metric = "lift", min_threshold = 1)
```

```python
result_association.sort_values('lift', ascending=False).head(20)
```
